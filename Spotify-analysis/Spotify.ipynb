# Spotify Data Analysis 

```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create the DataFrame

df_tracks = pd.read_csv('Data/tracks.csv')

# Import the first 5 rows of the dataset

df_tracks.head()
```

```python
# Check for null values

pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()

# Hence there are 71 missing values present in the name column
```

```python
# Top 10 list of least popular songs

sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
# Statistic analysis of dataset

df_tracks.describe().transpose()
```

```python
# Top 10 most popular songs on Spotify

most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
# Changing the original release_date column into date-time format

df_tracks.set_index("release_date", inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# Returns the artist in the index location row 18

df_tracks[["artists"]].iloc[18]
```

```python
# Converting the duration in ms to seconds and also add a duration column

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))

# Remove old column

df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
# New duration in seconds

df_tracks.duration.head()
```

```python
# Visualisation

corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="viridis", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

We can see that the strong correlation between loudness and energy (0.8) so we can explore this further by plotting a linear regression model fit.

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
# 0.4% of total dataset is 2346 rows

print(len(sample_df))
```

```python
# Linear regression shows that theres a strong correlation between loudness and energy

plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy Correlation")
```

```python
# Regression plot for popularity vs acousticness

plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "c").set(title = "Loudness vs Energy Correlation")
```

Linear regression shows that as the acousticness decreases, the popularity decreases.

```python
# Convert the release_date to date-time format

df_tracks["dates"] = df_tracks.index.get_level_values("release_date")
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

We can see that there were not many songs before 1940 and as the years go by, the number of songs released per year increases linearly perhaps due to advancement of technology to produce these songs.


```python
total_dr = df_tracks.duration
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth= False).set(title="Year vs Duration")
plt.xticks(rotation=90)

```

Earlier in the 1900s, the duration of the songs are slightly lower compared to late 1940s, and the duration remained consistent until around 2013.

```python

total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```

Songs from 1920s to 1960s have comparatively shorter duration due to the fact that songs were singing based rather than instrumental. 
From 2010 onwards, theres a steep decline in the duration which may be associated our attention span.
